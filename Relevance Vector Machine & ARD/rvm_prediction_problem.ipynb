# Problems with predictive distribution of RVM

```python
import numpy as np
from rvm import SparseBayesianLearner
import matplotlib.pyplot as plt
%matplotlib inline

np.random.seed(0)
# create data set
X = np.zeros([200,1])
x = np.zeros([1000,1])
X[0:100,0]   = 0.1*np.random.random(100) - 10
X[100:200,0] = 0.1*np.random.random(100) + 10
Y            = X[:,0]**2 + np.random.normal(0,2,200)
x[:,0]       = np.linspace(-20,20,1000) 

# train rvm
rvm = SparseBayesianLearner(learn_type = "regression", 
                            method     = "fixed-point",
                            alpha_max  = 10,
                            kernel     = "gaussian",
                            scaler     = 10)
rvm.fit(X,Y)

# predict mean & variance of posterior distribution
mu,var = rvm.predictive_distribution(x)

# plot variance 
plt.figure(figsize = (12,8))
plt.plot(x[:,0],var)
plt.xlabel("x")
plt.ylabel("variance of predictive distribution")
plt.title("Variance of predictive distribution for each x in test set")
plt.show()
plt.figure(figsize = (12,10))
plt.xlim((-20,20))
plt.hist(X[:,0])
plt.xlabel("x")
plt.ylabel("number of data points")
plt.title("Distribution of explanatory variable in training set")
plt.show()




```
