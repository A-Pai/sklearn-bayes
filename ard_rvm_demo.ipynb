## ARD regression demo

### Variable Selection

```python
import numpy as np
from sparse_bayesian_learner import SparseBayesianLearner
from sklearn.cross_validation import train_test_split
import matplotlib.pyplot as plt
%matplotlib inline

# generate training & test data
X                             = np.random.random([500,100])
X[:,0]                        = np.linspace(start = 0,stop = 10, num = 500)
Y                             = 10*X[:,0]  + np.random.normal(0,1,500)
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 0.2)

# fit data with EM & fixed-point
ard_regression_em             = SparseBayesianLearner(X_train, Y_train, method = "EM",alpha_max = 1)
ard_regression_fp             = SparseBayesianLearner(X_train, Y_train, method = "fixed-point",alpha_max = 1)
ard_regression_em.fit()
ard_regression_fp.fit()

# plot variance of parameters in prior
fp = plt.plot(np.log(1./ard_regression_fp.diagA),"bo")
plt.xlabel("feature index")
plt.ylabel("log of variance of feature in prior")
plt.title("ARD regression with fixed-point fitter")
plt.plot(np.zeros(100),"r-", linewidth = 2)
plt.show()

em = plt.plot(np.log(1./ard_regression_em.diagA),"bo")
plt.xlabel("feature index")
plt.ylabel("log of variance of feature in prior")
plt.title("ARD regression with EM fitter")
plt.plot(np.zeros(100),"r-", linewidth = 2)
plt.show()

print "number of variables selected wiht EM fitter " + str(np.sum(ard_regression_em.active))
print "number of variables selected wiht fixed-point fitter " + str(np.sum(ard_regression_fp.active))
```

```python
em_vars_selected = []
fp_vars_selected = []
for alpha_max_i in xrange(1,100):
     ard_regression_em  = SparseBayesianLearner(X_train, Y_train, method = "EM",alpha_max = alpha_max_i)
     ard_regression_fp  = SparseBayesianLearner(X_train, Y_train, method = "fixed-point",alpha_max = alpha_max_i)
     ard_regression_em.fit()
     ard_regression_fp.fit()
     em_vars_selected.append(np.sum(ard_regression_em.active))
     fp_vars_selected.append(np.sum(ard_regression_fp.active))
        
plt.plot(sorted(1./np.linspace(1,100,99)), em_vars_selected,"r-", linewidth = 2, label = "EM")
plt.plot(sorted(1./np.linspace(1,100,99)), fp_vars_selected,"b-", linewidth = 2, label = "Fixed Point")
plt.xlabel("threshold variance")
plt.ylabel("number of variables selected")
plt.legend(loc = 4)
plt.show()
```

### Predictions

```python
# predict posterior mean & variance ( using both EM & fixed-point fitted models)
mu_em,var_em                  = ard_regression_em.predict(X_test)
mu_fp,var_fp                  = ard_regression_fp.predict(X_test)
```
