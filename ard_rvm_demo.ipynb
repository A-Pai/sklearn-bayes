## ARD regression demo

### Variable Selection

```python
import numpy as np
from sparse_bayesian_learner import SparseBayesianLearner
from sklearn.cross_validation import train_test_split
import matplotlib.pyplot as plt
%matplotlib inline

# generate training & test data
X                             = np.random.random([500,100])
X[:,0]                        = np.linspace(start = 0,stop = 10, num = 500)
Y                             = 10*X[:,0]  + np.random.normal(0,1,500)
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 0.2)

# fit data with EM & fixed-point
ard_regression_em             = SparseBayesianLearner(method = "EM",alpha_max = 1)
ard_regression_fp             = SparseBayesianLearner(method = "fixed-point",alpha_max = 1)
ard_regression_em.fit(X_train,Y_train)
ard_regression_fp.fit(X_train,Y_train)

# plot variance of parameters in prior
fp = plt.plot(np.log(1./ard_regression_fp.diagA),"bo")
plt.xlabel("feature index")
plt.ylabel("log of variance of feature in prior")
plt.title("ARD regression with fixed-point fitter")
plt.plot(np.zeros(100),"r-", linewidth = 2)
plt.show()

em = plt.plot(np.log(1./ard_regression_em.diagA),"bo")
plt.xlabel("feature index")
plt.ylabel("log of variance of feature in prior")
plt.title("ARD regression with EM fitter")
plt.plot(np.zeros(100),"r-", linewidth = 2)
plt.show()

print "number of variables selected wiht EM fitter " + str(np.sum(ard_regression_em.active))
print "number of variables selected wiht fixed-point fitter " + str(np.sum(ard_regression_fp.active))
```

```python
em_vars_selected = []
fp_vars_selected = []
for alpha_max_i in xrange(1,100):
     ard_regression_em_i  = SparseBayesianLearner( method = "EM",alpha_max = alpha_max_i)
     ard_regression_fp_i  = SparseBayesianLearner(method = "fixed-point",alpha_max = alpha_max_i)
     ard_regression_em_i.fit(X_train, Y_train, )
     ard_regression_fp_i.fit(X_train, Y_train, )
     em_vars_selected.append(np.sum(ard_regression_em_i.active))
     fp_vars_selected.append(np.sum(ard_regression_fp_i.active))
        
plt.plot(np.log(sorted(1./np.linspace(1,100,99))), em_vars_selected,"ro", 
                                                                    markersize = 10, 
                                                                    label = "EM")
plt.plot(np.log(sorted(1./np.linspace(1,100,99))), fp_vars_selected,"b+", 
                                                                    markersize = 10, 
                                                                    label = "Fixed Point")
plt.ylim((-5,50))
plt.xlabel("log of threshold variance")
plt.ylabel("number of variables selected")
plt.legend(loc = 4)
plt.show()
```

```python
    # generate data
    x                               = np.ones([1000,1])
    x[:,0]                          = np.linspace(start = -5, stop = 5, num = 1000)
    y                               = 10*np.sinc(x[:,0]) + np.random.normal(0,1,1000)
    x_train,x_test, y_train, y_test = train_test_split(x,y, test_size = 0.5)
    
    # fit rvm regression model
    rvmKernelised   = SparseBayesianLearner(method = "fixed-point", kernel = "gaussian", scaler = 1)
    rvmKernelised.fit(x_train,y_train,)
    
    # fit svm regression model
    from sklearn.svm import SVR
    from sklearn.grid_search import GridSearchCV
    svr = GridSearchCV( SVR(kernel = 'rbf'), param_grid = {"gamma": np.logspace(-2,2,5),"C":[0.01,1,10,100,1000]}, cv=5)
    svr.fit(x_train,y_train)
    
    # predict test set with rvm & svm
    mu, var   = rvmKernelised.predict(x_test)
    svr_predict = svr.predict(x_test)
    
    print "MSE of RVM is {0}, number of support vectors is {1}".format(np.mean((mu-y_test)**2), 
                                                                       np.sum(rvmKernelised.active))
    print "MSE of SVM is {0}, number of support vectors is {1}".format(np.mean((svr_predict - y_test)**2), 
                                                                       svr.best_estimator_.support_vectors_.shape[0])
    
    # plot predicted vs data
    plt.plot(x_test[:,0],y_test,'ro', markersize = 6, label = "data")
    plt.plot(x_test[:,0],mu,'bd', label = "predicted values", linewidth = 3)
    plt.title("RVM, predicted values vs test data")
    plt.legend(loc = 1)
    plt.show()
    
    # plot support vectors
    plt.plot(x_test[:,0],mu,'b+', label = "predicted values")
    plt.plot(rvmKernelised.support_vecs[:,0],y_train[rvmKernelised.active],"rD",label = "support vectors")
    plt.title("RVM, predicted values and support vectors")
    plt.legend(loc = 1)
    plt.show()
    
    
```
