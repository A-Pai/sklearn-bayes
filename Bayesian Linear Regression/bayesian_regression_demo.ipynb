# Bayesian Regression

## Example

### Variance of coefficients

```python
import os
os.chdir("/Users/amazaspshaumyan/Desktop/Bayesian Regression Methods/Bayesian Linear Regression")
import numpy as np
from sklearn.linear_model import LinearRegression
from bayesian_regression import BayesianRegression
import matplotlib.pyplot as plt
from sklearn.cross_validation import train_test_split
%matplotlib inline

# generate features that will be highly correlated
np.random.seed(0)
n, m              = 100, 30
prototype         = 5*np.random.random(100)
X                 = np.outer(prototype,np.ones(m)) + 0.05*np.random.randn(n, m)

# Define only one feature to be relevant
Theta             = np.zeros(m)
Theta[0]          = 4.

# Ganerate linear model with noise
Y                 = np.dot(X, Theta) + np.random.normal(0,1,100)

# train & test split
X,x,Y,y           = train_test_split(X,Y, test_size = 0.4)

# Fit the Bayesian Regression and an OLS for comparison
br                = BayesianRegression(X,Y, bias_term = True)
br.fit()

ols               = LinearRegression()
ols.fit(X, Y)

plt.plot(ols.coef_,"r-",label = "OLS")
plt.plot(br.w_mu  ,"b-",label = "Bayesian")
plt.plot(Theta,"g-", label = "Truth", linewidth = 3)
plt.xlabel("parameter index")
plt.ylabel("parameter estimarte")
plt.title("OLS vs Bayesian Regression")
plt.legend(loc = 1)
plt.show()

```

### Point Prediction 

```python
# train data
Mu        = br.predict(X) # mean of predictive distribution
Y_hat     = ols.predict(X)

# train mse
mse_ols_train   = np.mean((Y_hat - Y)**2)
mse_bayes_train = np.mean((Mu    - Y)**2)

# test data
mu        = br.predict(x) # mean of predictive distribution
y_hat     = ols.predict(x)

# test mse
mse_ols_test   = np.mean((y_hat - y)**2)
mse_bayes_test = np.mean((mu    - y)**2)

print "Training set \n"
print "MSE for Bayesian Regression {0} on training set".format(mse_bayes_train)
print "MSE for OLS {0} on training set \n".format(mse_ols_train)

print "Test set \n"
print "MSE for Bayesian Regression {0} on test set".format(mse_bayes_test)
print "MSE for OLS {0} on test set".format(mse_ols_test)
```

### Predictive Distribution

```python
x          = np.zeros([100,1])
x[:,0]     = np.linspace(-3,3,100)
y          = 2 + 2*x[:,0] + np.random.normal(0,1,100)
br2        = BayesianRegression(x,y,bias_term = True)
br2.fit()
y_hat, var = br2.predict_dist(x)

plt.figure(figsize = (12,8))
plt.plot(x[:,0],y,"b+", label = "data")
plt.plot(x[:,0],y_hat,"ro", label = "mean prediction")
plt.plot(x[:,0],y_hat - np.sqrt(var),"g-",label = "means +- standard deviation")
plt.plot(x[:,0],y_hat + np.sqrt(var),"g-",label = "means +- standard deviation")
plt.legend(loc = 2)
plt.xlabel("x")
plt.ylabel("y")
plt.title("predictive distribution, Bayesian Linear Regression")
plt.show()
```
