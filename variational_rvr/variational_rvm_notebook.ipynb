```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cross_validation import train_test_split
from sklearn.metrics import mean_squared_error as mse
from vrvm import VRVR
from sklearn.datasets import load_boston
from sklearn.svm import SVR
from sklearn.grid_search import GridSearchCV
from sklearn.preprocessing import scale
import warnings
from vrvm import VariationalRegressionARD, VRVR
from fast_scikit_rvm import RegressionARD, RVR
%matplotlib inline



from scipy import stats
# Parameters of the example
n_samples, n_features = 300, 210
# Create Gaussian data
np.random.seed(0)
X = np.random.randn(n_samples, n_features)

# Create weigts
lambda_ = 4.
w = np.zeros(n_features)

# select 
relevant_features = np.random.randint(0, n_features, 20)
for i in relevant_features:
   w[i] = stats.norm.rvs(loc=0, scale=1. / np.sqrt(lambda_))
    
alpha_ = 1.
noise = stats.norm.rvs(loc=0, scale=1. / np.sqrt(alpha_), size=n_samples)
    
# Create the target
y = np.dot(X, w) + noise
X,x,Y,y = train_test_split(X,y, test_size = 0.8)

ard  = RegressionARD()
vard = VariationalRegressionARD()

ard.fit(X,Y)
vard.fit(X,Y)

print "ARD mse = {0}".format(mse(y,ard.predict(x)))
print "Variational ARD mse = {0}".format(mse(y,vard.predict(x)))
```

### Synthetic example

```python
# parameters
n = [20,40,60,80,100,150,200,250,300,400,500,750,1000,2000,5000]
n_sim      = 100
test_train = 0.7

# discard all warnings (for small number of observations there are
# sometimes cases when all features are pruned)
warnings.filterwarnings('ignore',".*")

def comparison(n):
    '''
    Comparing RVR, VRVR and SVR 
    '''
    # generate data set
    Xc       = np.ones([n,1])
    Xc[:,0]  = np.linspace(-5,5,n)
    Yc       = 10*np.sinc(Xc[:,0]) + np.random.normal(0,1,n)
    X,x,Y,y  = train_test_split(Xc,Yc,test_size = test_train, random_state = 0)

    # train rvm with fixed-point optimization
    rvr = SparseBayesianLearner(learn_type = "regression",method="fixed-point", alpha_max  = 10,
                                                                                kernel     = "gaussian",
                                                                                scaler     = 1)
    rvr.fit(X,Y)
    y_rvr,var_rvr = rvr.predictive_distribution(x)
    rvr_err       = mse(y_rvr,y)
    rvr_rvs       = np.sum(rvr.active[1:])
    
    # train variational rvm
    vrvr       = VRVR(X,Y,kernel = "rbf", scaler = 1, prune_thresh = 1e-2)
    vrvr.fit()
    y_vrvr,var_vrvr = vrvr.predict_dist(x)
    vrvr_err        = mse(y_vrvr,y)
    vrvr_rvs        = np.sum(vrvr.active[1:])

    # train svr
    svr        = GridSearchCV(SVR(kernel = "rbf"), param_grid = {'gamma':np.logspace(-2,2,5),
                                                             'C':np.logspace(-2,2,5)},
                                                   cv = 3)
    svr.fit(X,Y)
    y_svr      = svr.predict(x)
    svr_err    = mse(y_svr,y)
    svr_svs    = np.shape(svr.best_estimator_.support_vectors_)[0]
    
    return {'rvr':[rvr_err,rvr_rvs,var_rvr],'vrvr':[vrvr_err,vrvr_rvs,var_vrvr],
            'svr':[svr_err,svr_svs]}

mse_rvr,mse_svr,mse_vrvr = [],[],[]
rvs_rvr,rvs_svr,rvs_vrvr = [],[],[]
var_rvr,var_vrvr         = [],[]

for n_obs in n:
    r_mse,v_mse,s_mse = 0.,0.,0.
    r_rvs,v_rvs,s_rvs = 0.,0.,0.
    r_var,v_var       = 0.,0.
    for i in range(n_sim):
        result = comparison(n_obs)
        r_mse += result['rvr'][0]; r_rvs += result['rvr'][1]
        r_var += np.mean(result['rvr'][2])
        v_mse += result['vrvr'][0]; v_rvs += result['vrvr'][1]
        v_var += np.mean(result['vrvr'][2])
        s_mse += result['svr'][0]; s_rvs += result['svr'][1]
    mse_rvr.append(r_mse/n_sim); rvs_rvr.append(r_rvs/n_sim)
    mse_vrvr.append(v_mse/n_sim); rvs_vrvr.append(v_rvs/n_sim)
    mse_svr.append(s_mse/n_sim); rvs_svr.append(s_rvs/n_sim)
    var_rvr.append(r_var/n_sim); var_vrvr.append(v_var/n_sim)
```

```python
# comparing accuracy of prediction on test set
plt.figure(figsize = (6,9))
plt.plot(n,mse_rvr,"r-",label = 'rvm')
plt.plot(n,mse_vrvr,'b-',label = 'vrvm')
plt.plot(n,mse_svr,'g-',label = 'svr')
plt.title('N observations vs MSE')
plt.xlabel('N')
plt.ylabel('MSE')
plt.legend()
plt.show()
```

```python
# comparing sparsity of models
plt.figure(figsize = (6,9))
plt.plot(n,np.log(rvs_rvr),"r-",label = 'rvm')
plt.plot(n,np.log(rvs_vrvr),'b-',label = 'vrvm')
plt.plot(n,np.log(rvs_svr),'g-',label = 'svr')
plt.xlabel('N')
plt.ylabel('log of number of support / relevant vectors')
plt.title('N observations vs sparsity')
plt.legend()
plt.show()
```

```python
# comparing estimate of variance : vrvr vs rvr
plt.figure(figsize = (6,9))
plt.plot(n,var_rvr,'r-',label = 'RVR')
plt.plot(n,var_vrvr,'b-',label = "VRVR")
plt.xlabel('N')
plt.ylim((0,4))
plt.ylabel('Variance')
plt.title('N observations vs Variance Estimate')
plt.legend()
plt.show()
```

### Boston housing example

```python
# number of runs
n_sim       = 100

# load data , split training & test
boston      = load_boston()
XXb,YYb     = scale(boston["data"]),boston["target"]

# MSE & Relevant Vectors
mse_rvr,mse_vrvr,mse_svr = [],[],[]
rvs_rvr,rvs_vrvr,rvs_svr = [],[],[]
var_rvr,var_vrvr         = [],[]

for i in range(n_sim):
    Xb,xb,Yb,yb = train_test_split(XXb,YYb,test_size = 0.3)
    
    # RVM
    rvrb = SparseBayesianLearner(method = "fixed-point",learn_type = "regression",alpha_max=1e+3,
                                                                                  scaler = 10,
                                                                                  kernel="poly",
                                                                                  p_order = 3)
    rvrb.fit(Xb,Yb)
    y_rvrb,var_rvrb   = rvrb.predictive_distribution(xb)
    rvrb_err          = mse(y_rvrb,yb)
    rvrb_rvs          = np.sum(rvrb.active[1:])
    mse_rvr.append(rvrb_err); rvs_rvr.append(rvrb_rvs)
    var_rvr.append(np.mean(var_rvrb))

    # VRVM
    vrvrb             = VRVR(Xb,Yb,kernel = "poly", scaler = 10, prune_thresh = 1e-4, order = 3)
    vrvrb.fit()
    y_vrvr,var_vrvrb  = vrvrb.predict_dist(xb)
    vrvrb_err         = mse(y_vrvr,yb)
    vrvrb_rvs         = np.sum(vrvrb.active[1:])
    mse_vrvr.append(vrvrb_err); rvs_vrvr.append(vrvrb_rvs)
    var_vrvr.append(np.mean(var_vrvrb))
    
    # SVR ('gamma' can't be larger than 1, SVR hangs )
    svr               = GridSearchCV(SVR(kernel='poly',coef0=1,degree=3),
                                     param_grid = {"gamma":[0.01,0.02,0.05,0.08,0.001,0.005,0.008],
                                                  "C":np.logspace(-3,3,7)},
                                    cv = 3)
    svr.fit(Xb,Yb)
    y_svr             = svr.predict(xb)
    svr_err           = mse(y_svr,yb)
    svr_rvs           = np.shape(svr.best_estimator_.support_vectors_)[0]
    mse_svr.append(svr_err); rvs_svr.append(svr_rvs)
    

print "VRVR median MSE = {0}, median number of RVs = {1}".format(np.median(mse_vrvr),np.median(rvs_vrvr))
print "RVR median MSE = {0}, median number of RVs = {1}".format(np.median(mse_rvr),np.median(rvs_rvr))
print "SVR median MSE = {0}, median number of SVs = {1}".format(np.median(mse_svr),np.median(rvs_svr))
```

```python
# comparing prediction accuracy on test set in terms of MSE
fig, ax1 =  plt.subplots(figsize=(10, 6))
bp       =  plt.boxplot((mse_vrvr,mse_rvr,mse_svr))
plt.setp(ax1, xticklabels=['VRVR','RVR','SVR'])
plt.title("MSE Comparison: VRVR  - vs -   RVR  - vs - SVR")
plt.xlabel("Model Type")
plt.ylabel("Mean Squared Error")
plt.show()
```

```python
# comparing model sparsity in terms of relevant and support vectors
fig, ax1 =  plt.subplots(figsize=(10, 6))
bp       =  plt.boxplot((rvs_vrvr,rvs_rvr,rvs_svr))
plt.setp(ax1, xticklabels=['VRVR','RVR','SVR'])
plt.title("MSE Comparison: VRVR  - vs -   RVR  - vs - SVR")
plt.xlabel("Model Type")
plt.title("Sparsity Comparison: VRVR  - vs -  RVR  - vs - SVR")
plt.ylabel("Mean Squared Error")
plt.ylim((0,500))
plt.show()
```
