# Bayesian GMM with Automatic Model Selection

In this notebook we demonstrate use of Variational Bayesian GMM with automatic model selection. Model that we use is not fully Bayesian, mixing probabilities are considered to be parameters not random variables, so we use VBEM to fit the model. VBEM algorithm consists of two steps: E step and M step.

   E step : approximate posterior of all random variables (including latent variable) using Variational Bayes and evaluate lower bound. 
   
   M step : maximize lower bound with respect to mixing probabilities.
           
In the process of iterations some weights are driven to zero and are pruned if they achieve threshold level (by default 1e-5). So this algorithm uses the same Automatic Relevance Determination technique that was used by Tipping in first version of RVM.
It is also worth noting that in original paper authors used only one iteration of Variational Bayes in E step in order to speed up algorithm. There is obvious trade off between accuracy of approximation (this requires more iterations of VB in E step) and speed and sometimes it might be better to allow more iterations to improve perfomance 
(to change number of VB iterations use n_mfa_iter parameter in our implementation).

```python
import numpy as np
import matplotlib.pyplot as plt
from skbayes.mixture_models import VBGMMARD
from matplotlib.patches import Ellipse
import pandas as pd
%matplotlib inline



def plotter(X, max_k,title, rand_state = 1, prune_thresh = 1e-5, mfa_max_iter = 10):
    '''
    Plotting function for VBGMMARD clustering
    
    Parameters:
    -----------
    X: numpy array of size [n_samples, n_features]
       Data matrix
       
    max_k: int
       Maximum number of components
       
    title: str
       Title of the plot
       
    Returns:
    --------
    :instance of VBGMMARD class 
    '''
    # fit model & get parameters
    gmm = VBGMMARD(n_components = max_k, prune_thresh = prune_thresh,
                   n_mfa_iter = mfa_max_iter)
    gmm.fit(X)
    centers = gmm.means_
    covars  = gmm.covars_
    k_selected = centers.shape[0]
    
    # plot data
    fig, ax = plt.subplots(figsize = (10,6))
    ax.plot(X[:,0],X[:,1],'bo', label = 'data')
    ax.plot(centers[:,0],centers[:,1],'rD', markersize = 8, label = 'means')
    for i in range(k_selected):
        plot_cov_ellipse(pos = centers[i,:], cov = covars[i], ax = ax)
    
    plt.xlabel('x1')
    plt.ylabel('x2')
    plt.legend(loc = 2)
    plt.title((title + ', {0} initial clusters, {1} selected clusters').format(max_k,k_selected))
    plt.show()
    return gmm
    
    
# plot_cov_ellipse function is taken from  
# https://github.com/joferkington/oost_paper_code/blob/master/error_ellipse.py

def plot_cov_ellipse(cov, pos, nstd=2, ax=None, **kwargs):
    """
    Plots an `nstd` sigma error ellipse based on the specified covariance
    matrix (`cov`). Additional keyword arguments are passed on to the 
    ellipse patch artist.

    Parameters
    ----------
        cov : The 2x2 covariance matrix to base the ellipse on
        pos : The location of the center of the ellipse. Expects a 2-element
            sequence of [x0, y0].
        nstd : The radius of the ellipse in numbers of standard deviations.
            Defaults to 2 standard deviations.
        ax : The axis that the ellipse will be plotted on. Defaults to the 
            current axis.
        Additional keyword arguments are pass on to the ellipse patch.

    Returns
    -------
        A matplotlib ellipse artist
    """
    def eigsorted(cov):
        vals, vecs = np.linalg.eigh(cov)
        order = vals.argsort()[::-1]
        return vals[order], vecs[:,order]

    if ax is None:
        ax = plt.gca()

    vals, vecs = eigsorted(cov)
    theta = np.degrees(np.arctan2(*vecs[:,0][::-1]))

    # Width and height are "full" widths, not radius
    width, height = 2 * nstd * np.sqrt(vals)
    ellip = Ellipse(xy=pos, width=width, height=height, angle=theta, fill = False,
                    edgecolor = 'k',linewidth = 4,**kwargs)

    ax.add_artist(ellip)
    return ellip
```

## Synthetic Data Example 1

Figure below shows that VBGMMARD algorithm can get stuck in local optimum (remember optimization problem is nonconvex). However you can see that found clusters are not completely unreasonable.

```python
X = np.zeros([600,2])
np.random.seed(0)
X[0:200,:]   = np.random.multivariate_normal(mean = (0,10), cov = [[3,0],[0,2]], size = 200)
X[200:400,:] = np.random.multivariate_normal(mean = (0,0) , cov = [[3,0],[0,2]], size = 200)
X[400:600,:] = np.random.multivariate_normal(mean = (0,-10) , cov = [[3,0],[0,2]], size = 200)

sy_gmm_1 = plotter(X,20,'Synthetic Example')
```

## Synthetic Data Example 2 

In this case algorithm correctly identified 3 out of 20 clusters (note all gaussian components have  the same mean with coordinates (0,0) but different covariance matrices)

```python
X = np.zeros([600,2])
np.random.seed(0)
X[0:200,:]   = np.random.multivariate_normal(mean = (0,0), cov = [[3,-2],[5,-8]], size = 200)
X[200:400,:] = np.random.multivariate_normal(mean = (0,0), cov = [[-10,10],[-10,10]], size = 200)
X[400:600,:] = np.random.multivariate_normal(mean = (0,0), cov = [[5,-1],[5,-1]], size = 200)

sy_gmm_2 = plotter(X,20,'Synthetic Example', prune_thresh = 1e-5, mfa_max_iter = 10)
```
