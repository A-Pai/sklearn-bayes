# Bayesian Bernoulli Mixture Model

This notebook illustrates use of Bayesian Bernoulli Mixture for image clustering. We used subset of data used in kaggle digit recognition competition that contains examples of three digits, namely '1','3','6'. 

```python
import numpy as np
import pandas as pd
from skbayes.mixture_models import VBBMM

# load data (you can find data file at:)
digits = np.array(pd.read_csv('digits.csv'))

bmm = VBBMM(n_components = 3)
bmm = bmm.fit(digits)

# succes probabilities
```

```python
bmm.means_.shape
```

```python
sd = np.reshape(bmm.means_[:,0],(28,28))
import matplotlib.pyplot as plt
plt.imshow(sd)
```
