# Variational Bayes Logistic Regression with Automatic Relevance Determination

VBClassificationARD is modified version of Variational Bayes Logistic Regression with Jaakola&Jordan local bound. Instead of spherical covariance that is used to model prior of coefficients in 
VB Logistic Regression VBClassificationARD uses diagonal covariance with separate precision parameters for each coefficient.

### What is difference between ClassificationARD and VBClassificationARD?

There are two main differences:
    - ClassificationARD uses Sparse Bayesian Learner algorithm
      to fit model, while VBClassificationARD uses 
      Variational Bayes

    - Both can not compute posterior of coefficients
      analytically, so they resort to approximations.     
      ClassificationARD uses Laplace approximation, which is 
      significantly faster but less accurate, while   
      VBClassificationARD relies on Jaakola & Jordan 
      approximation.
      
In general, it is advised to use VBClassificationARD with small datasets and/or in cases when the accurate approximation of posterior is essential in all other cases ClassificationARD should be preferred.

### VBClassificationARD passes all scikit-learn tests for linear classifier

```python
from sklearn.utils.estimator_checks import check_estimator
from skbayes.rvm_ard_models import VBClassificationARD
check_estimator(VBClassificationARD)
```

### Example:  Classification with many irrelevant features

We use binary classification problem to compare the performance of Regularised Logistic Regression (with L1 and L2 regularizations) with VBClassificationARD. Toy dataset that we use for testing contains 500 features, only two of them are relevant, and the rest is just random noise. The figure below visualizes dataset in space spanned by two relevant features. As you can see two classes are 'almost' linearly separable.

```python
def generate_dataset(n_samples = 500, n_features = 100,
                     cov_class_1 = [[1.4,.5],[1.,.75]],
                     cov_class_2 = [[0.9,0.1],[1.7,.2]],
                     mean_class_1 = (1,1),
                     mean_class_2 = (-1,-1)):
    ''' Generate binary classification problem with two relevant features'''
    X   = np.random.randn(n_samples, n_features)
    Y   = np.ones(n_samples)
    sep = int(n_samples/2)
    Y[0:sep]     = 0
    X[0:sep,0:2] = np.random.multivariate_normal(mean = mean_class_1, 
                   cov = cov_class_1, size = sep)
    X[sep:n_samples,0:2] = np.random.multivariate_normal(mean = mean_class_2,
                        cov = cov_class_2, size = n_samples - sep)
    return X,Y

X,Y = generate_dataset()
plt.figure(figsize = (8,6))
plt.plot(X[Y==0,0],X[Y==0,1],"bo", markersize = 3)
plt.plot(X[Y==1,0],X[Y==1,1],"ro", markersize = 3)
plt.xlabel('feature 1')
plt.ylabel('feature 2')
plt.title("Example of dataset")
plt.show()
```

To compare different algorithms we run 50 simulations in which we randomly generate dataset similar to one in the figure above and classify it using all three algorithms.
Obtained results show that VBClassificationARD on average has better accuracy than Logistic Regression with L1 / L2 penalty.

```python
import sklearn
from matplotlib import cm
import matplotlib
from sklearn.cross_validation import train_test_split
from sklearn.linear_model import LogisticRegressionCV

lrcverr_l2, lrcverr_l1, vblrerr = [],[],[]
n_sim = 50
n_samples, n_features = 700, 500


for i in range(n_sim): 
    # Parameters of the example
    X,Y = generate_dataset(n_samples,n_features)
    
    # training & test data
    X,x,Y,y = train_test_split(X,Y, test_size = 0.4)
    # scale data
    mux,stdx = np.mean(X,0),np.std(X,0)
    X = (X - mux) / stdx
    x = (x - mux) / stdx
    
    # fit & predict using ARD
    vblr = VBClassificationARD()
    vblr.fit(X,Y)
    y_vblr = vblr.predict(x)
        
    # fit & predict using cross validated Logistic Regression
    # with L1 / L2 penalty
    lrcv_l2 = LogisticRegressionCV(penalty = 'l2')
    lrcv_l2.fit(X,Y)
    y_lrcv_l2 = lrcv_l2.predict(x)
    
    lrcv_l1 = LogisticRegressionCV(penalty = 'l1', solver = 'liblinear')
    lrcv_l1.fit(X,Y)
    y_lrcv_l1 = lrcv_l1.predict(x)
    
    lrcverr_l1.append(np.sum(y!=y_lrcv_l1))
    lrcverr_l2.append(np.sum(y!=y_lrcv_l2))
    vblrerr.append(np.sum(y!=y_vblr))    

    
n = x.shape[0]
ard = 1 - np.asarray(vblrerr, dtype = np.float)/n
l1  = 1 - np.asarray(lrcverr_l1,dtype = np.float)/n
l2  = 1 - np.asarray(lrcverr_l2, dtype = np.float)/n 
better = np.sum((ard > l1) & (ard > l2)) * 100./ n_sim
print "============= Simulation Results =========================== \n"
print "Number of simulations = {0} \n".format(n_sim)
print "Average accuracy for VBClassificationARD {0} \n".format(np.mean(ard))
print "Average accuracy for L2 LogisticRegressionCV {0} \n".format(np.mean(l2))
print "Average accuracy for L1 LogisticRegressionCV {0} \n".format(np.mean(l1))
print ("VBClassificationARD had better accuracy than L1 / L2 Regularised "
       "Logistic Regression in {0} percents of simulations").format(better)
print "=============================================================\n"

plt.figure(figsize = (8,6))
plt.ylim((-.1,1.1))
plt.plot(ard,'b-', label = 'VBClassificationARD')
plt.plot(l1,'r-', label = 'L1')
plt.plot(l2,'g-',label = 'L2')
plt.xlabel('simulation index')
plt.ylabel('Accuracy')
plt.title("Accuracy comparison by simulation")
plt.legend(loc = 3)
plt.show()
```

### Example Continues: Predictive Distribution and Decision Boundary

Since we know generating process for each class (i.e. P(X|C1) and P(X|C2)), we can easily compute predictive distribution (i.e. P(C1|X) and P(C2|X)). The figure below compares predictive distribution obtained by each of three models with 'real' predictive distribution obtained through knowledge of the generating process for each class. 

```python
# Parameters for binary classification problem
from scipy.stats import multivariate_normal as mvn
n_samples = 500
n_features = 500
cov1 = [[.9,.1],[.6,.8]]
cov2 = [[1.4,.5],[.1,.5]]
mean1 = (-0.5,1.5)
mean2 = (1,-1.5)
np.random.seed(1)

X,Y = generate_dataset(n_features = n_features,
                       cov_class_1 = cov1,
                       cov_class_2 = cov2,
                       mean_class_1 = mean1,
                       mean_class_2 = mean2)

# training & test data
X,x,Y,y = train_test_split(X,Y, test_size = 0.4)

# fit & predict using ARD
vblr = VBClassificationARD()
vblr.fit(X,Y)
        
# fit & predict using cross validated Logistic Regression
# with L1 / L2 penalty
lrcv_l2 = LogisticRegressionCV(penalty = 'l2')
lrcv_l2.fit(X,Y)
lrcv_l1 = LogisticRegressionCV(penalty = 'l1', solver = 'liblinear')
lrcv_l1.fit(X,Y)
    
# construct grid    
n_grid = 100
max_x      = np.max(x[:,0:2],axis = 0)
min_x      = np.min(x[:,0:2],axis = 0)
X1         = np.linspace(min_x[0],max_x[0],n_grid)
X2         = np.linspace(min_x[1],max_x[1],n_grid)
x1,x2      = np.meshgrid(X1,X2)
Xgrid      = np.zeros([n_grid**2,2])
Xgrid[:,0] = np.reshape(x1,(n_grid**2,))
Xgrid[:,1] = np.reshape(x2,(n_grid**2,))
Xg         = np.random.randn(n_grid**2,n_features)
Xg[:,0]    = Xgrid[:,0]
Xg[:,1]    = Xgrid[:,1]

# estimate probabilities for grid data points
vblr_grid    = vblr.predict_proba(Xg)[:,1]
lrcv_l2_grid = lrcv_l2.predict_proba(Xg)[:,1]
lrcv_l1_grid = lrcv_l1.predict_proba(Xg)[:,1]

# compute 'real' predictive distribution
def real_posterior(x,mu1,mu2,cov1,cov2,prior1=0.5, prior2=0.5):
    lpf1 = mvn.logpdf(x,mu1,cov1) + np.log(prior1)
    lpf2 = mvn.logpdf(x,mu2,cov2) + np.log(prior2)
    log_prob = np.logaddexp(lpf1,lpf2)
    return np.exp(lpf1 - log_prob)
rp = real_posterior(Xg[:,0:2], mean1, mean2, cov1, cov2)

# plot data
titles = ["VBClassificationARD","LogisticRegressionCV with L2 penalty",
          "LogisticRegressionCV with L1 penalty"]
models  = [vblr_grid,lrcv_l2_grid,lrcv_l1_grid]

print "            Model ========================== vs ======================  Real  "

for title,model in zip(titles,models):
    fig, axarr = plt.subplots(nrows=1, ncols=2, figsize = (25,10))
    axarr[0].contourf(X1,X2,np.reshape(model,(n_grid,n_grid)),cmap=cm.coolwarm)
    axarr[1].contourf(X1,X2,np.reshape(rp,(n_grid,n_grid)),cmap=cm.coolwarm)
    axarr[0].plot(x[y==0,0],x[y==0,1],"bo", markersize = 5)
    axarr[1].plot(x[y==0,0],x[y==0,1],"bo", markersize = 5)
    axarr[0].plot(x[y==1,0],x[y==1,1],"ro", markersize = 5)
    axarr[1].plot(x[y==1,0],x[y==1,1],"ro", markersize = 5)
    axarr[0].set_title(title)
    axarr[1].set_title("Real Predictive Distribution")
    plt.show()
```

Notice also how choppy are decision boundaries of both versions of Logistic Regression, that happens because coefficients for irrelevant features are non-zero.
